## Jawaban nomor 2 Masukkan informasi di atas dalam dictionary dengan key sesuai nama informasinya

```python
Jumlah = len(df)
```

```python
nama = list(df.columns)
```

```python
jumlahObjek = len(df.select_dtypes(include=['object']).columns)
jumlahAngka = len(df.select_dtypes(include=['int64', 'float64']).columns)
jumlahNULL = len(df.columns[df.isna().sum() > 0])
```

```python
kamus = {
    "Jumlah Record":Jumlah,
    "Nama Atribut" : nama,
     "Jumlah Objek": jumlahObjek,
    "Jumlah Angka": jumlahAngka,
    "Jumlah NULL": jumlahNULL,
    
}
```

```python
A=pd.DataFrame(kamus)
print(tabulate(A,headers="keys",tablefmt="psql",showindex=False))
```
